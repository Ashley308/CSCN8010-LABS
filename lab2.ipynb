```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/Users/ranjan/Documents/conestoga /data Analysis mathematics/teenagers_social_media_dataset.csv')

# Print actual column names
print("Actual column names in the dataset:")
print(df.columns.tolist())

# Function to get the closest matching column name
def get_column(df, name):
    for col in df.columns:
        if name.lower().replace(" ", "") in col.lower().replace(" ", ""):
            return col
    return None

# Columns to analyze
numeric_cols = ['Age', 'Time Spent', 'Average Posts Per Day', 'Follower Count']
categorical_cols = ['Location', 'Devices Used', 'Purpose', 'Privacy Settings', 
                     'Cyberbullying', 'Perceived Impact on Mental Health', 
                     'Participation in Trends/Challenges', 'Content Preferences', 
                     'Parental Control']

# Display statistics for numeric columns
for col_name in numeric_cols:
    col = get_column(df, col_name)
    if col:
        print(f"\nStatistics for {col}:")
        print(f"Mean: {df[col].mean():.2f}")
        print(f"Median: {df[col].median():.2f}")
        print(f"Mode: {df[col].mode().values[0]:.2f}")
    else:
        print(f"\nColumn similar to '{col_name}' not found.")

# Display mode for categorical columns
for col_name in categorical_cols:
    col = get_column(df, col_name)
    if col:
        print(f"\nMode for {col}: {df[col].mode().values[0]}")
    else:
        print(f"\nColumn similar to '{col_name}' not found.")
```
